```python
tweets=pandas_data_handler.download_csv_to_dataframe(lookups.CsvUrlTweets.ELON_MUSK_TWEETS)
```

```python

# Assuming you have a DataFrame called 'tweets'

# Filter rows where 'Text' column contains the word 'dodge' (case-insensitive)
tweets = tweets[tweets['Text'].str.contains('doge', case=False, na=False)]

# Rename the 'Text' column to 'tweet'
tweets.rename(columns={'Text': 'tweet'}, inplace=True)


```

```python
tweets
```

```python
from textblob import TextBlob

# Assuming you have a DataFrame called 'tweets' with a 'tweet' column
tweets['sentiment'] = tweets['tweet'].apply(lambda tweet: TextBlob(tweet).sentiment.polarity)
```

```python
# Assuming you have a DataFrame called 'tweets' with a 'sentiment' column

# Define a function to categorize sentiment
def categorize_sentiment(sentiment):
    if sentiment > 0:
        return 'positive'
    elif sentiment < 0:
        return 'negative'
    else:
        return 'neutral'

# Apply the categorize_sentiment function to create a new column 'sentiment_category'
tweets['sentiment_category'] = tweets['sentiment'].apply(categorize_sentiment)
```

```python
# Create a copy of the 'tweets' DataFrame with only positive sentiment
positive_tweets = tweets[tweets['sentiment_category'] == 'positive'].copy()
```
